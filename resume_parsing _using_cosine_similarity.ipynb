```python
import re
import pdfplumber

with pdfplumber.open("SiddakMResume2.pdf") as pdf:
    first_page = pdf.pages[0]
    first_page_text = first_page.extract_text()

    # Define the starting and ending words
    start_word = "EXPERIENCE"
    end_word = "PROJECT"
    
    # Find the indices of the starting and ending words
    start_index = first_page_text.find(start_word)
    end_index = first_page_text.find(end_word)

    if start_index != -1 and end_index != -1:
        # Extract the text between the starting and ending words
        extracted_text = first_page_text[start_index:end_index]

        # Print the extracted text
        print(extracted_text)
    else:
        print("Start or end word not found.")
```

```python
lines = [line.strip() for line in extracted_text.split('\n') if line.strip()]

# Print the array of lines
print(lines)
```

```python
import spacy

# Load spaCy model with GloVe embeddings
nlp = spacy.load("en_core_web_md")

# Example extracted lines

# Function to convert text lines to dense vectors
def lines_to_vectors(lines):
    vectors = []
    for line in lines:
        # Process text with spaCy
        doc = nlp(line)

        # Get the average vector for the document
        dense_vector = doc.vector

        vectors.append(dense_vector)

    return vectors

# Convert lines to dense vectors
dense_vectors_resume = lines_to_vectors(lines)

# Print the dense vectors
for vector in dense_vectors_resume:
    print(vector)
```
